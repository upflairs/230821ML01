```python
df1.groupby('Year')[['BasePay','OvertimePay']].mean()
```

```python
df1.groupby('Year')[['BasePay','OvertimePay']].sum()
```

```python
df1.groupby('Year')[['BasePay','OvertimePay']].std()
```

```python
df1.groupby('Year')[['BasePay','OvertimePay']].max()
```

```python
df1.groupby('Year')[['BasePay','OvertimePay']].min()
```

```python
df1.groupby('Year')[['BasePay','OvertimePay']].agg([np.mean, max, min, len])
```
