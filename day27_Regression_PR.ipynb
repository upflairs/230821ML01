```python
xtrain.shape
```

```python
model_linear = LinearRegression().fit(xtrain,ytrain)
```

# Polynomial Fit

```python
from sklearn import preprocessing as skpp
```

```python
x.shape
```

```python
x2 = xdf[['in']].values
x2.shape
```

```python
x2
```

```python
x_poly1 = skpp.PolynomialFeatures(1).fit_transform(x2)
```

```python
x_poly1
```

```python
x_poly2 = skpp.PolynomialFeatures(2).fit_transform(x2)
x_poly2
```

```python
x_poly3 = skpp.PolynomialFeatures(3).fit_transform(x2)
x_poly3
```

```python
model_poly1 = LinearRegression().fit(x_poly1, y)
```

```python
model_poly1.score(x_poly1,y)
```

```python
model_poly2 = LinearRegression().fit(x_poly2, y)
model_poly2.score(x_poly2,y)
```

```python
plt.figure(figsize=(12,6))
plt.scatter(x2,y,c='r')
plt.scatter(x2,model_poly2.predict(x_poly2), c='b')
plt.legend(['Origianl Data','Polynomial Fitting'])
```

```python
model_poly3 = LinearRegression().fit(x_poly3, y)
model_poly3.score(x_poly3,y)
```

```python
plt.figure(figsize=(12,6))
plt.scatter(x2,y,c='r')
plt.scatter(x2,model_poly3.predict(x_poly3), c='b')
plt.legend(['Origianl Data','Polynomial Fitting'])
```

```python
scores = []
for i in range(1, 10):
    poly = skpp.PolynomialFeatures(i).fit_transform(x2)
    model_poly_i = LinearRegression().fit(poly, y)
    scores.append(model_poly_i.score(poly,y))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,10),scores, marker='o')
```

# Startup Profit Prediction

```python
data = pd.get_dummies(data)
data.head()
```

```python
model_linear = LinearRegression().fit(data.drop(['Profit','Administration'],axis=1), data['Profit'])
model_linear.score(data.drop(['Profit','Administration'],axis=1), data['Profit'])
```

```python
model_linear2 = LinearRegression().fit(data.drop(['Profit'],axis=1), data['Profit'])
model_linear2.score(data.drop(['Profit'],axis=1), data['Profit'])
```

```python
rnd = pd.DataFrame(skpp.PolynomialFeatures(3).fit_transform(data[['R&D Spend']]), columns='rnd0,rnd1,rnd2,rnd3'.split(','))
mark = pd.DataFrame(skpp.PolynomialFeatures(3).fit_transform(data[['Marketing Spend']]), columns='mark0,mark1,mark2,mark3'.split(','))
```

```python
rnd
```

```python
poly_df = data[['State_California','State_Florida','State_New York']]
```

```python
poly_df = pd.concat([poly_df,rnd,mark],axis=1)
```

```python
poly_df.head()
```

```python
model_poly = LinearRegression().fit(poly_df, data['Profit'])
model_poly.score(poly_df, data['Profit'])
```
