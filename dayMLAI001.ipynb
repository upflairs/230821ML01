```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('target',axis=1)
y = df['target']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3, random_state=100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier().fit(xtrain, ytrain)
```

```python
model.score(xtrain,ytrain)*100
```

```python
model.score(xtest,ytest)*100
```

```python
# Elbow Method
scores = []
for i in range(2,30):
    model_i = KNeighborsClassifier(n_neighbors=i).fit(xtrain, ytrain)
    scores.append(model_i.score(xtest,ytest)*100)
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(2,30),scores, marker='o',markersize=10,lw=3,markerfacecolor='r')
plt.xticks(range(2,30))
plt.ylim([0,100])
plt.yticks(range(0,101,5))
plt.xlabel('Number of Neighbors')
plt.ylabel('Score')
plt.show()
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest, model.predict(xtest)))
```
