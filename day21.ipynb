### Merge

```python
gargi = pd.DataFrame({'roll':[3,7,10,109,38,44,55,67],
                      'name':['Rahul','Vinay','Saroj','Manish','Vineeta','Akshay','Rajkumar','Abhishek'],
                      'branch':['ec','ec','cs','cs','cs','ec','cs','ee']})
neha = pd.DataFrame({'roll':[3,7,10,38,44,45,46,100,109],
                     'place':['Jaipur','Jaipur','Jaipur','Jaipur','Jaipur','Ajmer','Jodhpur','Jaipur','Kota'],
                     'mobile':[9999888777,123456789,9876543,3456789,2345678,123456789,98765,2345678,9876]})
```

```python
gargi
```

```python
neha
```

```python
pd.merge(gargi,neha,on='roll')
```

```python
pd.merge(gargi,neha,on='roll',how='outer')
```

```python
pd.merge(gargi,neha,on='roll',how='left')
```

```python
mm = pd.merge(gargi,neha,on='roll',how='right')
mm
```
