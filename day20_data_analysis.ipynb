### Iteration over DataFrames

```python
for i in df:
    print(i)
```

```python
for i in df:
    print(df[i])
```

```python
for i in df.values:
    print(i)
    break
```

* iteritems
* iterrows
* itertuples

```python
for i in df.iteritems():
    print(i)
    break
```

```python
for ind, val in df.iteritems():
    print(ind)
    print(val)
    break
```

```python
for i in df.iterrows():
    print(i)
```

```python
for ind,row in df.iterrows():
    print(ind)
    print(row)
    print(row['type'])
    break
```

```python
for i in df.itertuples():
    print(i)
```

```python
for i in df.itertuples():
    print(i)
    break
```
