```python
x = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms','Area Population']]  # Features
y = df['Price'] # Labels
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.3, random_state=50)
```

#### Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
model_linear = LinearRegression().fit(xtrain, ytrain)
```

```python
pd.DataFrame(model_linear.coef_, index=x.columns, columns=['Coeff'])
```

#### Predictions

```python
model_linear.predict([[40000,6,5,30000]])
```

```python
model_linear.predict([[40000,6,5,30000]])[0]
```

#### Model Evaluation

```python
ypred = model_linear.predict(xtest)
```

```python
# Finding out Errors
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
print('MAE:',round(mean_absolute_error(ytest,ypred),2))
print('MSE:',round(mean_squared_error(ytest,ypred),2))
```

```python
# Finding the R2-Score
print(round(model_linear.score(xtest,ytest)*100, 2),'%')
```
