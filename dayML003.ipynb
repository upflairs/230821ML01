```python
amazon = 'https://www.amazon.in/s?k={}'
```

```python
amazon.format('jeans+for+men')
```

```python
r = requests.get(amazon.format('jeans+for+men'), headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36'})
```

```python
r.status_code
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
soup.find('span',{'class':'a-size-base-plus a-color-base a-text-normal'})
```

```python
titles = [i.text for i in soup.find_all('span',{'class':'a-size-base-plus a-color-base a-text-normal'})]
titles
```
