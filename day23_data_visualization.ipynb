```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips.info()
```

```python
plt.plot(tips['total_bill'])
```

```python
tips['total_bill'].plot()
```

```python
plt.figure(figsize=(20,8))
_ = plt.plot(tips['total_bill'],color='r',lw=2,ls='-',marker='*',markersize=15,markerfacecolor='yellow',markeredgecolor='blue')
_ = plt.title('Total Bill Line Plot')
_ = plt.grid()
_ = plt.xlabel('Entries-->')
_ = plt.ylabel('Bill Amount (in $)-->')
_ = plt.xticks(range(0,245,5))
```

```python
plt.figure(figsize=(20,8))
plt.subplot(1,2,1)
_ = plt.plot(tips['total_bill'],color='r',lw=2,ls='-',marker='*',markersize=15,markerfacecolor='yellow',markeredgecolor='blue')
_ = plt.title('Total Bill Line Plot')
_ = plt.grid()
_ = plt.xlabel('Entries-->')
_ = plt.ylabel('Bill Amount (in $)-->')
_ = plt.xticks(range(0,245,5))

plt.subplot(1,2,2)
_ = plt.plot(tips['tip'],color='g',lw=2,ls='-',marker='*',markersize=15,markerfacecolor='yellow',markeredgecolor='blue')
_ = plt.title('Total Tip Line Plot')
_ = plt.grid()
_ = plt.xlabel('Entries-->')
_ = plt.ylabel('Tip Amount (in $)-->')
_ = plt.xticks(range(0,245,5))
```
